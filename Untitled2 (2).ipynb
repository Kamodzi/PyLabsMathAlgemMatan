```python
x = sp.Symbol('x')
y1 = x**2 + 1
y2 = 3 - x
y3 = 1

S1 = sp.integrate(y1-y3, (x, 0, 1))
print(f'S 1: {S1}')

S2 = sp.integrate(y2-y3, (x, 1, 2))
print(f'S 2: {S2}')

print(f"Sum S = S1+S2 = {S1 + S2}")

y1 = lambda x: x**2 + 1
y2 = lambda x: 3 - x
y3 = lambda x: 1 + 0*x

x = np.linspace(-1, 2.2, 1000)
plt.plot(x, y1(x))
plt.plot(x, y2(x))
plt.plot(x, y3(x))

x = np.linspace(0, 1, 1000)
plt.fill_between(x=x, y1=np.minimum(y1(x), y2(x)), y2=y3(x), color='red', 
                 alpha=.7)

x = np.linspace(1, 2, 1000)
plt.fill_between(x=x, y1=np.minimum(y1(x), y2(x)), y2=y3(x), color='black', 
                 alpha=.7)
plt.grid()
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
