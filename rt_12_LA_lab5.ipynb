### Упражнение 5.2

```python
# Точка (-2/3, -2/3, -2/3) лежит в плоскости x + 2y + 3z + 4 = 0
# тогда
p = sp.Plane( (-2/3, -2/3, -2/3), normal_vector=(1, 2, 3) )
perp = p.perpendicular_plane( (4, 2, 3), (2, 0, 1) )
print(f"Уравнение перпендикулярной плоскости:\n{perp.equation()}")
```

### Упражнение 5.4

```python
p = sp.Plane( (1, -1, -3), normal_vector=(2, -3, 5) )
print(f"Уравнение плоскости: {p.equation()}")
print(f"Координаты нормального вектора: (2, -3, 5)")

print(f"A = 2\nB = -3\nC = 5\nD = 10")
A, B, C, D = 2, -3, 5, 10

p2 = sp.Plane( (0, 0, 0), normal_vector=(2, -3, 5) )
print(f"Расстояние от начала координат: {p.distance(p2)}")

f = lambda x, y: (-A * x - B * y - D) / C

fig = plt.figure(figsize=(10, 10))
ax  = fig.add_subplot(1, 1, 1, projection='3d')

xval = np.linspace(-4, 4, 100)
yval = np.linspace(-4, 4, 100)
x, y = np.meshgrid(xval, yval)

p = np.array([1, -1, -3])
v = np.array([2, -3, 5])
ax.quiver(*p, *v, color='b')

z = f(x, y)
surf = ax.plot_surface(x, y, z, rstride=10, cstride = 10, cmap=cm.plasma)
surf.set_alpha(0.5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Плоскость 2x - 3y + 5z + 10 = 0')

ax.plot([0], [0], [0], 'o', color='k')
ax.plot([1], [-1], [-3], 'o', color='c')

ax.view_init(30, 70)
plt.show()
```

### Упражнение 5.5

```python
M = np.array([2, 3, -4])
N = np.array([-3, 1, 3])

a = np.array([-3, 2, 1])
b = np.array([0, 3, 1])
n = np.cross(a, b)

p = sp.Plane( M, normal_vector=n )
print(f"Уравнение плоскости: {p.equation()}")
print(f"Нормальный вектор: {n}")

print(f"A = -1\nB = 3\nC = -9\nD = -43")
A, B, C, D = -1, 3, -9, -43

p2 = sp.Plane( N, normal_vector=n )
print(f"Расстояние от точки N(-3, 1, 3): {p.distance(p2)}")

f = lambda x, y: (-A * x - B * y - D) / C

fig = plt.figure(figsize=(10, 10))
ax  = fig.add_subplot(1, 1, 1, projection='3d')

_range = 6
xval = np.linspace(-_range, _range, 100)
yval = np.linspace(-_range, _range, 100)
x, y = np.meshgrid(xval, yval)

ax.quiver(*M, *n, color='b')

z = f(x, y)
surf = ax.plot_surface(x, y, z, rstride=10, cstride = 10, cmap=cm.plasma)
surf.set_alpha(0.5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Плоскость -x + 3y - 9z - 43 = 0')

ax.plot(*N, 'o', color='r')
ax.plot(*M, 'o', color='c')

ax.view_init(0, 70)
plt.show()
```

### Дополнительное задание

```python
def dist(L1: sp.Line, L2: sp.Line):
    A = L1.points[0]                # Точка А принадлежит L1
    n = L1.points[1] - L1.points[0] # Направляющий вектор прямой L1 будет
                                    # являться нормалью плоскости Р
        
    P = sp.Plane(A, normal_vector=n)       # Плоскость перп. к L1
    PL2_point = P.projection(L2.points[0]) # Проекция точки из L2 на Р
    PL2 = sp.Line(PL2_point, direction_ratio=L2.direction) # Проекция L2 на Р
    
    return PL2.distance(A)
    

line1 = sp.Line( point1, direction_ratio=(3, 4, 2) )
line2 = sp.Line( (7, 1, 3), direction_ratio=(3, 4, 2) )
print(dist(line1, line2))
```
