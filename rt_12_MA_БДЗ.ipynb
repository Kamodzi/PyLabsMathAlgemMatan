```python
import numpy as np
import matplotlib.pyplot as plt
```

### $y=x^2$ Ð¸ $y=\sqrt{x}$

```python
x = np.linspace(-4, 4, 1000)
y = x**2

x_inv = x[x >= 0]
y_inv = np.sqrt(x_inv)

plt.plot(x, y)[0].set_label("$y = x^2$")
plt.plot(x_inv, y_inv)[0].set_label("$y = \sqrt{x}$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

### $y = \sqrt[3]{x}$ Ð¸ $y = x^3$

```python
x = np.linspace(-2, 2, 1000)
y = np.cbrt(x)

x_inv = x
y_inv = (x_inv)**3

plt.plot(x, y)[0].set_label("$y = \sqrt[3]{x}$")
plt.plot(x_inv, y_inv)[0].set_label("$y = x^3$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

### $y=e^x$ Ð¸ $y = \ln{x}$

```python
x = np.linspace(-2, 2, 1000)
y = np.exp(x)

x_inv = x[x >= 0]
y_inv = np.log(x_inv)

plt.plot(x, y)[0].set_label("$y = e^x$")
plt.plot(x_inv, y_inv)[0].set_label("$y = \ln{x}$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

### $y=\log_{\frac{1}{2}}{x}$ Ð¸ $y=(\frac{1}{2})^x$

```python
x = np.linspace(-5, 5, 1000)
y = np.log(x[x >= 0]) / np.log(1/2)

x_inv = x
y_inv = (1/2)**x

plt.plot(x[x >= 0], y)[0].set_label("$y=\log_{\\frac{1}{2}}{x}$")
plt.plot(x_inv, y_inv)[0].set_label("$y=(\\frac{1}{2})^x$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

### $y=\sin{x}$ Ð¸ $y=\arcsin{x}$

```python
x = np.linspace(-5, 5, 1000)
y = np.sin(x)

x_inv = x[np.abs(x)<=1]
y_inv = np.arcsin(x_inv)

plt.plot(x, y)[0].set_label("$y=\sin{x}$")
plt.plot(x_inv, y_inv)[0].set_label("$y=\\arcsin{x}$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

### $y=\arctan{x}$ Ð¸ $y=\tan{x}$

```python
x = np.linspace(-10, 10, 1000)
y = np.arctan(x)

x_inv = x
y_inv = np.tan(x_inv)

plt.plot(x, y)[0].set_label("$y=\\arctan{x}$")
plt.plot(x_inv, y_inv)[0].set_label("$y=\\tan{x}$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 2)
plt.grid()
plt.legend()
plt.show()
```

### $y=\arccos{x}$ Ð¸ $y=\cos{x}$

```python
x = np.linspace(-10, 10, 1000)
y = np.arccos(x[np.abs(x)<1])

x_inv = x
y_inv = np.cos(x_inv)

plt.plot(x[np.abs(x)<1], y)[0].set_label("$y=\\arccos{x}$")
plt.plot(x_inv, y_inv)[0].set_label("$y=\\cos{x}$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

### $y=ctg {x}$ Ð¸ $y=arcctg {x}$

```python
x = np.linspace(-10, 10, 1000)
y = 1/np.tan(x)

x_inv = x
y_inv = (np.pi/2) - np.arctan(x_inv)

plt.plot(x, y)[0].set_label("$y=ctg{x}$")
plt.plot(x_inv, y_inv)[0].set_label("$y=arcctg{x}$")

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 5)
plt.grid()
plt.legend()
plt.show()
```
