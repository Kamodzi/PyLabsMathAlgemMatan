```python
print('C1(1)')


import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-3,1,20)
handle1, = plt.plot(x,np.sqrt(x+3),':go',label='$\sqrt{x+3}$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(handles=[handle1])
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k')
plt.title('Функция $\sqrt{x+3}$')
plt.show()

print('C1(2)')


import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-1,1,1000)
handle1, = plt.plot(x,np.sign(x),'--r*',label='$sign(x)$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(handles=[handle1])
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k')
plt.title('Функция $sign(x)$')
plt.show()


print("C2")

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2,2,20)
y = plt.plot(x,np.exp(x),'--g*',label='$sign(x)$')



x = np.linspace(np.exp(-2),np.exp(2),20)
y = plt.plot(x,np.log(x),':rd',label='$sign(x)$')

plt.xlabel('x')
plt.ylabel('y')
x = np.linspace(-2,np.exp(2),20)
y = plt.plot(x,x,'-.ks',label='$sign(x)$')

plt.grid(True)
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k')
plt.show()


```
