### Упражнение 1

#### $y_1=\dfrac{2x-1}{x-1}$

```python
x = np.linspace(1e10, 1e10+50, 1000)
y1 = (2*x-1) / (x-1)

plt.title("$\lim_{x->+\infty} y_1(x)$")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x, y1)
plt.show()
```

### Упражнение 2

#### a)  $\lim_{x\to+\infty} (1+x)^{\frac{1}{x}} $

```python
x = np.linspace(1e10, 1e10+50, 100, dtype=np.float64)
y = (1+x)**(1/x)

plt.title("$\lim_{x->+\infty} (1+x)^{1/x}$")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x, y)
plt.show()
```

####  $\lim_{x\to+\infty} (1+x)^{\frac{1}{x}} = 1 $

#### б)  $\lim_{x\to0} \dfrac{\sin{\dfrac{x^2}{\sqrt{1-x^3}}}}{x\ln{(1-x^3)}} $

```python
x = np.linspace(-0.5, 0.5, 1000, dtype=np.float64)
y = ( np.sin( (x**2)/(np.sqrt(1-x**3)) ) )/(x*np.log(1-x**3))

plt.title("$\lim_{x->0} \dfrac{\sin{\dfrac{x^2}{\sqrt{1-x^3}}}}{x\ln{(1-x^3)}} $")
plt.axvline(x=0, color='gray', ls='--')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x, y)
plt.show()
```

#### б)  $\lim_{x\to0} \dfrac{\sin{\dfrac{x^2}{\sqrt{1-x^3}}}}{x\ln{(1-x^3)}} = -\infty$

### Упражнение 3

```python
x = sp.Symbol('x')
y = (1+x)**(1/x)
lim = sp.limit(y, x, sp.oo)

print(f"Предел первой функции: {lim}")

x = sp.Symbol('x')
y = ( sp.sin( (x**2)/(sp.sqrt(1-x**3)) ) )/(x*sp.log(1-x**3))
lim = sp.limit(y, x, 0)

print(f"Предел второй функции: {lim}")
print("Вычисленные пределы соответствуют полученным приближенно")
```
