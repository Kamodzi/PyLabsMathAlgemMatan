```python
import numpy as np
import matplotlib.pyplot as plt
```

### Упражнение 1

```python
def sums(a, n0, eq_axis=False, size=4):
    if not hasattr(a, "__len__"):
    
        a = [a]
     
    
    
    for idx, series in enumerate(a):
        sums = [ series(1) ]
        for i in range(2, n0):
            sums.append( sums[-1] + series(i) )

        indexes = np.arange( 1, len(sums)+1, 1 )

        plt.plot( indexes, sums, 'o')

        plt.minorticks_on()
        plt.grid(which='major')
        plt.grid(which='minor', ls=':', lw=0.5)
        plt.xlabel('$n$')
        plt.ylabel('$S_n$')
        if eq_axis:
            plt.axis('equal')
    plt.show()
    
```

```python
def element(a, n0, eq_axis=False, size=5):
    if not hasattr(a, "__len__"):
        a = [a]
    
    
    
    for i, series in enumerate(a):    
        indexes = np.arange( 1, n0+1, 1 )
        elements = series(indexes)

        plt.plot( indexes, elements, 'o'
                 , ms=size)

        plt.minorticks_on()
        plt.grid(which='major')
        plt.grid(which='minor', ls=':', lw=0.5)
        plt.xlabel('$n$')
        plt.ylabel('$S_n$')
        
        
        
        if eq_axis:
            plt.axis('equal')
    plt.show()

a = lambda n: 1/n**2
sums(a, 1000)
```

### Упражнение 7

```python
def calc(a, e):
    n = 1
    summ = 0
    delta = a(n)
    while delta > e:
        summ += delta
        n += 1
        delta = a(n)
    return summ



a = lambda n: 1/n**2
print(calc(a, 0.001))
```

### Упражнение 8

```python
a = lambda n: n/(2**(n-1))

summ = calc(a,0.001)
print(f"Полученное значение: {summ}")
print("В таблице  4")
print(f"Разница: {np.abs(summ - 4)}")
```

### Упражнение 9

```python
def calcabs(a, e):
    n = 1
    summ = 0
    delta = a(n)
    while abs(delta) > e:
        summ += delta
        n += 1
        delta = a(n)
    return summ
```
