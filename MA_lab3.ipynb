### Упражнение 4

```python
z = [complex(1, 3), complex(3, 4)]
z.append(z[0] + z[1])
z.append(z[0] - z[1])

vectors = []
for el in z:
    vectors.append([np.real(el), np.imag(el)])

vectors = np.array(vectors)

plt.title("Сложение и вычитание комплексных чисел")
plt.axis('equal')
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')

origin = np.array([[0, 0, 0, 0],[0, 0, 0, 0]])
plt.quiver(*origin, vectors[:,0], vectors[:,1], color=['r','b','g','c'], scale=20)
plt.show()

print("Сложение комплексных чисел подобно сложению векторов")
```
