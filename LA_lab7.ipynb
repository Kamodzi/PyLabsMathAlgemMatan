```python
import numpy as np
import sympy as sp 
import matplotlib.pyplot as plt
```

### Упражнение 7.1

```python
e = sp.Ellipse(sp.Point(3, 1), hradius=3, eccentricity=sp.Rational(4, 5))

print(f"Параметры эллипса:")
print(f"Большая полуось:\t{e.hradius}")
print(f"Малая полуось:\t{e.vradius}")
print(f"Эксцентриситет:\t{e.eccentricity}")
print(f"Координаты фокусов:\t{e.foci}")
print(f"Фокусное расстояние:\t{e.focus_distance}")
print(f"Уравнение:\t{e.equation()}")
print(f"Параметрическое уравнение:\t{e.arbitrary_point()}")
print(f"Площадь:\t{e.area}")

t = np.arange(0, 2*np.pi, 0.01)

x = 3*np.cos(t) + 3
y = 9/5*np.sin(t) + 1
plt.plot(x, y, lw=2)[0].set_label("Ellipse")

f = e.foci
plt.plot(f[0].x, f[0].y, 'ro')[0].set_label("Focus 1")
plt.plot(f[1].x, f[1].y, 'go')[0].set_label("Focus 2")

plt.plot(3, 1, 'yo')[0].set_label("Center")

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.axis('equal')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```

### Упражнение 7.2

```python
p1 = sp.Parabola(sp.Point(0, 0), sp.Line(sp.Point(0, 8), sp.Point(1, 8)))

print(f"Параметры параболы:")
print(f"Параметр p:\t{p1.p_parameter}")
print(f"Фокус:\t{p1.focus}")
print(f"Фокусное расстояние:\t{p1.focal_length}")
print(f"Эксцентриситет:\t{p1.eccentricity}")
print(f"Директриса:\t{p1.directrix}")
print(f"Ось симметрии:\t{p1.axis_of_symmetry}")
print(f"Уравнение:\t{p1.equation()}")
print(f"y = -x**2/16 + 4")

x = np.linspace(-20, 20, 100)
y = -x**2/16 + 4
plt.plot(x, y)[0].set_label("Парабола")

plt.plot(p1.focus.x, p1.focus.y, 'ro')[0].set_label("Фокус")
plt.axhline(y=8, color='k').set_label("Директриса")

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.axis('equal')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

### Упражнение 7.3

```python
a = np.sqrt(2)
b = np.sqrt(8)

t = np.linspace(-2, 2, 1000)

x = a*np.cosh(t)
y = b*np.sinh(t)

plt.plot(x, y, 'k')[0].set_label("Гипербола")
plt.plot(-x, y, 'k')

x = np.linspace(-6, 6)
y = b/a*x
plt.plot(x, y)[0].set_label("Асимптота 1")
plt.plot(x, -y)[0].set_label("Асимптота 2")

c = np.sqrt(a**2+b**2)
plt.plot(c, 0, 'ro')[0].set_label("Фокус 1")
plt.plot(-c, 0, 'go')[0].set_label("Фокус 2")

plt.ylabel('$y$')
plt.xlabel('$x$')
# plt.axis('equal')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
