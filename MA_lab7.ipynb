```python
import numpy as np
import sympy as sp
from scipy import optimize
import matplotlib.pyplot as plt
from sympy.solvers import nsolve
from sympy import Symbol
```

### Упражнение 1

#### a)

```python
def y1(x):
    return (x**2-4)/(np.abs(x**2-4))

x = np.linspace(-10, 10, 1000)
y = y1(x)
plt.plot(x, y)

x1 = optimize.brentq(y1, -2.5, -1)
print(f"x1 = {x1}\nf(x1) = {y1(x1)}")

x2 = optimize.brentq(y1, 1, 2.5)
print(f"\nx1 = {x2}\nf(x1) = {y1(x2)}")

print("\nТочки разрыва: x = +- 2")
print("Разрыв первого рода")

plt.grid()
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

#### б)

```python
def y2(x):
    return np.exp(1/(x-2))

x = np.linspace(-3, 10, 1000)
y = y2(x)
plt.plot(x, y)

print("Разрыв второго рода в точке х = 2")

plt.grid()
plt.ylim(-0.25, 4)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

### Упражнение 2

#### a)

```python
def f(x):
    return 3 + np.sqrt(x) - np.exp(x/2)

x = np.linspace(0, 10, 1000)
y = f(x)
plt.plot(x, y)

x1 = optimize.brentq(f, 2, 4)
print(f"x1 = {x1}\nf(x1) = {f(x1)}")

plt.grid()
plt.ylim(-0.25, 4)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

#### б)

```python
def f(x):
    return x**3 - 8*x**2 + 17*x + np.sqrt(x) - 10

x = np.linspace(0, 6, 1000)
y = f(x)
plt.plot(x, y)

x1 = optimize.brentq(f, 0.5, 1)
print(f"x1 = {x1}\nf(x1) = {f(x1)}")

x2 = optimize.brentq(f, 2, 3)
print(f"\nx2 = {x2}\nf(x2) = {f(x2)}")

x3 = optimize.brentq(f, 4, 5)
print(f"\nx3 = {x3}\nf(x3) = {f(x3)}")

plt.grid()
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

### Упражнение 3

#### а)

```python
x = sp.Symbol('x')
eq = (3*x)**np.e - np.e**(3*x)
a = nsolve(eq, 0)

print(f"x = {a}")
print(f"f(x) = {eq.subs(x, a)}")
```

#### б)

```python
x = sp.Symbol('x')
eq = x/3 - sp.sin(x)
a = nsolve(eq, 0)

print(f"x = {a}")
print(f"f(x) = {eq.subs(x, a)}")
```

### Упражнение 4

```python
w, x, y, z = sp.symbols('w x y z')

f1 = w + x + 4*y + 3*z - 5
f2 = 2*w + 3*x + y - 2*z - 1
f3 = w + 2*x - 5*y + 4*z - 3
f4 = w -3*z - 9

a = nsolve((f1, f2, f3, f4), (w, x, y, z), (0, 0, 0, 0))
print(f"x = {a}")
```

### Упражнение С1

#### a)

```python
def y3(x):
    return np.arcsin(x)/x

x = np.linspace(-1, 1, 1000)
y = y3(x)
plt.plot(x, y)

print("\nТочка разрыва: x = 0")
print("Устраняемый разрыв")

plt.grid()
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

#### б)

```python
def y4(x):
    return 2/(x**2 - 1)

x = np.linspace(-7, 7, 1000)
y = y4(x)
plt.plot(x, y)

x1 = optimize.brentq(y4, -2, 0)
print(f"x1 = {x1}\nf(x1) = {y4(x1)}")

x2 = optimize.brentq(y4, 0, 2)
print(f"\nx2 = {x1}\nf(x2) = {y4(x2)}")

print("\nТочки разрыва: x = +-1")
print("Разрыв второго рода")

plt.grid()
plt.ylim(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

### Упражнение С2

```python
def f(x):
    return x**2 - 5*x*np.sin(3*x) + 3

x = np.linspace(-10, 10, 1000)
y = f(x)

x1 = optimize.brentq(f, -5, -2.5)
print(f"x1 = {x1}\nf(x1) = {f(x1)}")

x2 = optimize.brentq(f, -2.5, 1)
print(f"\nx2 = {x2}\nf(x2) = {f(x2)}")

x3 = optimize.brentq(f, 1, 2.5)
print(f"\nx3 = {x3}\nf(x3) = {f(x3)}")

x4 = optimize.brentq(f, 2.5, 5)
print(f"\nx4 = {x4}\nf(x4) = {f(x4)}")

plt.plot(x, y)
plt.grid()
plt.ylim(-5, 15)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

### Упражнение С3

#### а)

```python
x = sp.Symbol('x')
eq = x + 1/x - 2
a = nsolve(eq, 1)

print(f"x = {a}")
print(f"f(x) = {eq.subs(x, a)}")
```

#### б)

```python
x = sp.Symbol('x')
eq = sp.log(1+x) + sp.log(1-x**2)

a1 = nsolve(eq, 0)
a2 = nsolve(eq, 0.5)

print(f"x1 = {a1}")
print(f"f(x) = {eq.subs(x, a1)}")

print(f"\nx2 = {a2}")
print(f"f(x) = {eq.subs(x, a2)}")
```
